## Draw accuracy and loss function over epochs

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
PATH = './logs/log_7'
FILE = 'train_log.csv'
pathCSV = os.path.join(PATH, FILE)
```

```python
# read csv file
data = pd.read_csv(pathCSV)
```

```python
# Convert data to DataFrame
df = pd.DataFrame(data)

# Plotting Loss
plt.figure(figsize=(14, 7))
plt.plot(df['epoch'], df['loss'], label='Loss', color='r')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Over Epochs')
plt.legend()
plt.grid(True)
pathPNG = os.path.join(PATH, 'loss.png')
plt.savefig(pathPNG)
plt.show()
```

```python
# Convert data to DataFrame
df2 = pd.DataFrame(data)

# Plotting Categorical Accuracy
plt.figure(figsize=(14, 7))
plt.plot(df2['epoch'], df2['categorical_accuracy'], label='Categorical Accuracy', color='b')
plt.xlabel('Epoch')
plt.ylabel('Categorical Accuracy')
plt.title('Categorical Accuracy Over Epochs')
plt.legend()
plt.grid(True)
pathPNG = os.path.join(PATH, 'accuracy.png')
plt.savefig(pathPNG)
plt.show()
```
