### Dataset features

```python
import constants as cs
import globali as gg
import config_loader as cl
import matplotlib.pyplot as plt
import json
import os
```

```python
def init():
    action_videos = {}
    with open(cs.JSON_PATH, 'r') as file:
        data = json.load(file)    
    for action in data:
        new_list = []
        for instance in action['instances']:
            new_list.append(instance['video_id'])
        action_videos[action['gloss']] = new_list
    return os.listdir(cs.VIDEOS_PATH), action_videos
```

```python
# initialization
gg.actions, gg.counters = cl.read_yaml_fields()
```

```python
def count_videos():
    count = 0
    for action in gg.counters:
        count += gg.counters[action]
    return count

def average_for_action():
    return count_videos()/len(gg.actions)

def show_instagram():
    keys = list(gg.counters.keys())
    values = list(gg.counters.values())
    plt.bar(keys, values)
    
    plt.xlabel('Actions')
    plt.ylabel('instances')
    plt.title('Action Istogram')
    plt.show()
    
```

```python
print("Number actions: ", len(gg.actions))
print("Number videos: ", count_videos())
print("Average for action:", average_for_action())
```
